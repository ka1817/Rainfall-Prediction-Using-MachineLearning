```python
import pandas as pd
from scipy.stats import ttest_ind, mannwhitneyu, shapiro

# Split the data
rain_pressure = df[df['rainfall'] == 'yes']['pressure']
no_rain_pressure = df[df['rainfall'] == 'no']['pressure']

# Shapiro-Wilk test for normality
shapiro_rain = shapiro(rain_pressure)
shapiro_no_rain = shapiro(no_rain_pressure)

normal_rain = shapiro_rain.pvalue > 0.05
normal_no_rain = shapiro_no_rain.pvalue > 0.05

print("Normality Test Results:")
print(f"Rainfall = 1: p-value = {shapiro_rain.pvalue:.4f} → {'Normal' if normal_rain else 'Not normal'}")
print(f"Rainfall = 0: p-value = {shapiro_no_rain.pvalue:.4f} → {'Normal' if normal_no_rain else 'Not normal'}")

# Choose test based on normality
if normal_rain and normal_no_rain:
    t_stat, p_val = ttest_ind(rain_pressure, no_rain_pressure, equal_var=False)
    print("\nT-test Results:")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_val:.4f}")
else:
    u_stat, p_val = mannwhitneyu(rain_pressure, no_rain_pressure, alternative='two-sided')
    print("\nMann-Whitney U Test Results:")
    print(f"U-statistic: {u_stat:.4f}")
    print(f"P-value: {p_val:.4f}")

# Interpretation
if p_val < 0.05:
    print("✅ Statistically significant difference in Pressure between Rainfall and No Rainfall.")
else:
    print("❌ No statistically significant difference in Pressure between Rainfall and No Rainfall.")
```

📊 Visual Insight (KDE Plot):


The pressure distributions for rainfall = 'yes' and 'no' are overlapping, but not identical.

rainfall = yes appears slightly more concentrated around 1010 hPa, with a sharper peak.

rainfall = no has a broader distribution, extending further to higher pressures (around 1030+ hPa).

This suggests that lower pressures may be more associated with rainfall, which aligns with meteorological intuition.

🧪 Statistical Insight (Mann-Whitney U Test):


Normality test failed (p-values < 0.05 for both groups): ✅ Correct to use Mann-Whitney U.

U-statistic = 13039, p-value = 0.1057.

Since p > 0.05, the difference is not statistically significant at the 95% confidence level.

❌ Conclusion: We cannot reject the null hypothesis — i.e., there's no strong statistical evidence that pressure distributions are different between rainfall and no-rainfall groups.

🧠 Overall Interpretation:


Visually, there’s a pattern suggesting slightly lower pressures during rainfall, but...

Statistically, the difference is not strong enough to be significant in this dataset.

```python
import pandas as pd
from scipy.stats import ttest_ind, mannwhitneyu, shapiro

# List of numeric features to test
features = ['pressure', 'maxtemp', 'temparature', 'mintemp', 'dewpoint',
            'humidity', 'cloud', 'sunshine', 'winddirection', 'windspeed']

results = []

for feature in features:
    group_yes = df[df['rainfall'] == 'yes'][feature]
    group_no = df[df['rainfall'] == 'no'][feature]

    # Normality Test
    shapiro_yes = shapiro(group_yes)
    shapiro_no = shapiro(group_no)

    normal_yes = shapiro_yes.pvalue > 0.05
    normal_no = shapiro_no.pvalue > 0.05

    # Select appropriate test
    if normal_yes and normal_no:
        test_type = 'T-test'
        stat, p_val = ttest_ind(group_yes, group_no, equal_var=False)
    else:
        test_type = 'Mann-Whitney U'
        stat, p_val = mannwhitneyu(group_yes, group_no, alternative='two-sided')

    # Significance interpretation
    significant = p_val < 0.05

    results.append({
        'Feature': feature,
        'Shapiro Rain p': shapiro_yes.pvalue,
        'Shapiro NoRain p': shapiro_no.pvalue,
        'Normal Rain': normal_yes,
        'Normal NoRain': normal_no,
        'Test': test_type,
        'Statistic': stat,
        'P-value': p_val,
        'Significant': '✅ Yes' if significant else '❌ No'
    })

# Convert to DataFrame for display
results_df = pd.DataFrame(results)
print("\n📊 Statistical Test Results by Feature:\n")
print(results_df[['Feature', 'Test', 'P-value', 'Significant']])
```
